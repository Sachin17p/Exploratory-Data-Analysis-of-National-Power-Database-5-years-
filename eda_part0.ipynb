```python
import pandas as pd
```

```python
df = pd.read_csv('test_data.csv')
```

```python
df
```

```python
import matplotlib.pyplot as plt
plt.rcParams["figure.dpi"] = 300
```

```python
plt.plot(df['course'].to_numpy(), df['participants'].to_numpy(),'*')
plt.xlabel('Course')
plt.ylabel('Participants')
```

```python
plt.bar(df['course'], df['participants'])
plt.xlabel('Course')
plt.ylabel('Participants')
```

```python
plt.pie(df['participants'], labels=df['course'], autopct='%1.2f%%');
plt.savefig('figures/test_data_pie.jpg', dpi=300,bbox_inches='tight')
```
